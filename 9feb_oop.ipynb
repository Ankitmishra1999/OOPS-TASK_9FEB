```python
Q.3 What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

ans.

class Vehicle:
    def __init__(self, name_of_vehicle, max_speed):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed

    def display_info(self):
        return f"Vehicle: {self.name_of_vehicle}, Max Speed: {self.max_speed}"

class Electric:
    def __init__(self, battery_capacity, charging_time):
        self.battery_capacity = battery_capacity
        self.charging_time = charging_time

    def display_info(self):
        return f"Battery Capacity: {self.battery_capacity}, Charging Time: {self.charging_time}"

class ElectricCar(Vehicle, Electric):
    def __init__(self, name_of_vehicle, max_speed, battery_capacity, charging_time):
        Vehicle.__init__(self, name_of_vehicle, max_speed)
        Electric.__init__(self, battery_capacity, charging_time)

    def display_info(self):
        vehicle_info = Vehicle.display_info(self)
        electric_info = Electric.display_info(self)
        return f"{vehicle_info}, {electric_info}"

# Creating an instance of ElectricCar
my_electric_car = ElectricCar("Tesla Model S", 250, "75 kWh", "8 hours")

# Accessing the display_info method from ElectricCar class (demonstrating method resolution order)
print(my_electric_car.display_info())
```

```python
Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.

ans.

Getter method: A getter is a method that retrieves the value of a private attribute.
It provides read-only access to the attribute.

Setter method: A setter is a method that sets the value of a private attribute. 
It provides a way to modify the attribute while enforcing any necessary validations or constraints.

Let's create a class called Person with a private attribute age,
and then create a getter and a setter for the age attribute:
    
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    # Getter for the age attribute
    def get_age(self):
        return self._age

    # Setter for the age attribute
    def set_age(self, new_age):
        if new_age < 0:
            raise ValueError("Age cannot be negative.")
        self._age = new_age

# Creating an instance of the Person class
person = Person("John", 30)

# Using the getter to retrieve the age
print("Current age:", person.get_age())

# Using the setter to change the age
person.set_age(35)

# Using the getter again to see the updated age
print("Updated age:", person.get_age())
    
```

```python
Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

ans.

class Animal:
    def make_sound(self):
        return "Generic animal sound"

class Dog(Animal):
    def make_sound(self):
        return "Bark bark!"

class Cat(Animal):
    def make_sound(self):
        return "Meow meow!"

# Creating instances of the subclasses
dog = Dog()
cat = Cat()

# Calling the overridden method for each subclass
print(dog.make_sound())  # Output: Bark bark!
print(cat.make_sound())  # Output: Meow meow!
```
